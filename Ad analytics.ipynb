```python
import pandas as pd
import numpy as np
```

```python
url = "https://gist.githubusercontent.com/andron23/f0ef92f6eae15b148f703fedf21121e7/raw/27be9b20a2eb668df8535d81fc628751ed1c553b/vk.csv"
vk = pd.read_csv(url, sep = ';', decimal = ',')
# decimal for floats
```

```python
url1 = "https://gist.githubusercontent.com/andron23/bae617f82d16691584e392136e054719/raw/468bf0879b362fabbb78ce05254b05c5943700de/inst.csv"
inst = pd.read_csv(url1, sep = ';', decimal = ',')
```

```python
inst.head()
```

```python
vk.info()
```

```python
vk.describe()
```

```python
# new column
vk['placement'] = 'vk'
inst['placement'] = 'inst'
```

```python
vk.head()
```

```python
# concatenate tables
df = pd.concat([vk, inst]).sort_values(['date', 'placement'])
```

```python
df.head(40)
```

# CTR, click through rate

```python
df['ctr'] = df['clicks']/df['impressions']
df['ctr'].head()
```

# CR1, conversion rate 1

```python
df['cr1'] = df['leads']/df['clicks']
df['cr1'].head()
```

# CPL - cost per lead

```python
df['cpl'] = df['spent']/df['leads']
df['cpl'].head()
```

# CPQL - cost per qualified lead

```python
df['cpql'] = df['spent']/df['qualleads']
df['cpql'].head()
```

# CR-Q - lead to a qualified lead

```python
df['cr-q'] = df['qualleads']/df['leads']
df.head()
```

# CPC - cost per click

```python
df['cpc'] = df['spent']/df['clicks']
df['cpc'].head()
```

# Cleaning

```python
df.replace([np.inf, -np.inf], np.nan, inplace = True)
```

```python
df.tail()
```

# Mean, Median

```python
df.groupby('placement').agg({
'ctr':'mean',
'cpl':'mean'})
```

```python
df2 = df.groupby('placement').agg({
    'clicks':[
        ('sum_clicks', 'sum'),
        ('mean_clicks', 'mean')
    ],
     'impressions':[
        ('sum_impressions', 'sum'),
        ('mean_impressions', 'mean')
    ],
     'spent':[
        ('total_spent', 'sum'),
        ('mean_spent', 'mean')
    ],
    'leads':[('cnt_leads', 'sum')],
    'qualleads':[('cnt_qual', 'sum')],
    'customers':[('cnt_customers', 'sum')]
}).reset_index()

df2.columns = df2.columns.droplevel(0)
```

```python
df2
```

```python
df2['crq'] = df2['cnt_qual']/df2['cnt_leads']
df2['cpql'] = df2['total_spent']/df2['cnt_qual']
df2['cpl'] = df2['total_spent']/df2['cnt_leads']
df2['cpc'] = df2['total_spent']/df2['sum_clicks']
df2['cr-1'] = df2['cnt_leads']/df2['sum_clicks']

df2['cr-2'] = df2['cnt_customers']/df2['cnt_leads']
df2['cac'] = df2['total_spent']/df2['cnt_customers']
```

```python
df2
```

# ARPU - total revenue divided by the number of users

```python
df2 = df.groupby('placement').agg({
    'clicks':[
        ('sum_clicks', 'sum'),
        ('mean_clicks', 'mean')
    ],
     'impressions':[
        ('sum_impressions', 'sum'),
        ('mean_impressions', 'mean')
    ],
     'spent':[
        ('total_spent', 'sum'),
        ('mean_spent', 'mean')
    ],
    'leads':[('cnt_leads', 'sum')],
    'qualleads':[('cnt_qual', 'sum')],
    'customers':[('cnt_customers', 'sum')],
    'rubs':[('sales', 'sum')]
}).reset_index()

df2.columns = df2.columns.droplevel(0)
```

```python
df2['crq'] = df2['cnt_qual']/df2['cnt_leads']
df2['cpql'] = df2['total_spent']/df2['cnt_qual']
df2['cpl'] = df2['total_spent']/df2['cnt_leads']
df2['cpc'] = df2['total_spent']/df2['sum_clicks']
df2['cr-1'] = df2['cnt_leads']/df2['sum_clicks']
df2['cr-2'] = df2['cnt_customers']/df2['cnt_leads']
df2['cac'] = df2['total_spent']/df2['cnt_customers']

df2['arpu'] = df2['sales']/df2['cnt_customers']
```
